```python
data[["participant", "abbreviation", "origin", "q_words", "?_words"]].sort_values(by="origin", ascending=False)
```

```python
statistics_df = pd.DataFrame(
    data[["participant", "abbreviation", "origin", "sex", "age", "education"]].copy()
)

statistics_df["q_total"] = data.filter(like="q_").sum(axis=1)
statistics_df["?_total"] = data.filter(like="?_").sum(axis=1)
statistics_df["total"] = statistics_df["q_total"] + statistics_df["?_total"]

# sort by origin
statistics_df.sort_values(by="origin", inplace=True, ignore_index=True)

statistics_df
```

```python
general_statistics = (
    statistics_df[["abbreviation", "origin", "q_total", "?_total", "total"]]
    .copy()
    .sort_values(by=["total"], ignore_index=True)
)

# make percentage representation for each allophone [q, ?]
general_statistics["% [q]"] = (
    (general_statistics["q_total"] / general_statistics["total"]) * 100
).round(2)
general_statistics["% [?]"] = (
    (general_statistics["?_total"] / general_statistics["total"]) * 100
).round(2)
general_statistics.drop(columns=["q_total", "?_total"], inplace=True)
general_statistics.reset_index(drop=True, inplace=True)
general_statistics
```

```python
mean = statistics_df["total"].mean()
median = statistics_df["total"].median()
print(f"Mean: {mean}. Median: {median}")

statistics_df
```

```python
hypothesis_1 = statistics_df[["abbreviation", "origin", "?_total", "q_total"]].copy()
# rename abbreviation to participant
hypothesis_1.rename(columns={"abbreviation": "participant"}, inplace=True)
hypothesis_1["allophone"] = hypothesis_1.apply(
    lambda row: "q" if row["q_total"] > row["?_total"] else "?", axis=1
)
hypothesis_1[["participant", "origin", "?_total", "q_total", "allophone"]].copy().rename(
    columns={
        "participant": "Participant",
        "?_total": "# [?]",
        "q_total": "# [q]",
        "allophone": "Favoured Allophone",
    },
)
```

```python
# Group the data by origin and sort participants within each group
hypothesis_1_sorted = hypothesis_1.sort_values(
    by=["origin", "participant"], ascending=[True, True]
)

# Calculate percentages for q and ? pronunciations
hypothesis_1_sorted["q_percentage"] = (
    (
        hypothesis_1_sorted["q_total"]
        / (hypothesis_1_sorted["q_total"] + hypothesis_1_sorted["?_total"])
    )
    * 100
).round(2)
hypothesis_1_sorted["?_percentage"] = (
    (
        hypothesis_1_sorted["?_total"]
        / (hypothesis_1_sorted["q_total"] + hypothesis_1_sorted["?_total"])
    )
    * 100
).round(2)

# Plot the bar chart with percentages
plt.figure(figsize=(12, 8))
plt.bar(
    hypothesis_1_sorted["participant"],
    hypothesis_1_sorted["q_percentage"],
    label="[q]",
    alpha=0.8,
)
plt.bar(
    hypothesis_1_sorted["participant"],
    hypothesis_1_sorted["?_percentage"],
    label="[?]",
    bottom=hypothesis_1_sorted["q_percentage"],
    alpha=0.8,
)

# Add labels and title
plt.xlabel("Participants (Origin noted below)")
plt.ylabel("Percentage Pronunciation")
plt.title("Percentage Pronunciation of [q] and [?] by Participant and Origin")

# Add origin as a note below each participant
x_labels = (
    hypothesis_1_sorted["participant"] + "\n(" + hypothesis_1_sorted["origin"] + ")"
)
plt.xticks(
    ticks=range(len(hypothesis_1_sorted)), labels=x_labels, rotation=45, ha="right"
)

# Add a legend for [q] and [?] only
plt.legend(title="Allophones", loc="upper right")

# Add y-axis scale points with a step of 10
plt.yticks(range(0, 101, 10))  # Scale points at 0, 10, 20, ..., 100

# Adjust layout and save the plot
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Filter data for participant 'A'
participant_a_data = data[data['abbreviation'] == 'A']

# Calculate the total counts for [q] and [?]
q_total = participant_a_data.filter(like='q_').sum(axis=1).values[0]
question_total = participant_a_data.filter(like='?_').sum(axis=1).values[0]

# Calculate percentages
total = q_total + question_total
q_percentage = (q_total / total) * 100
question_percentage = (question_total / total) * 100

# Plot the results as a pie chart
plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    [q_percentage, question_percentage],
    labels=None,  # Remove labels
    autopct='%1.1f%%',  # Display percentages on the chart
    startangle=90       # Rotate the pie chart for better orientation
)

# Add a legend
plt.legend(wedges, ['[q]', '[?]'], title="Allophones", loc="upper right")

# Add a title
plt.title("Participant A: Allophone Percentage Distribution")
plt.tight_layout()
plt.show()
```

```python
Q1 = statistics_df['total'].quantile(0.25)
Q3 = statistics_df['total'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
# Identify outliers
statistics_df['is_outlier'] = (statistics_df['total'] < lower_bound) | (statistics_df['total'] > upper_bound)

# Step 3: Create the box plot with individual points
plt.figure(figsize=(8, 6))

# Box plot to show distribution
sns.boxplot(y=statistics_df['total'], color='lightblue')

# Strip plot to overlay individual points, with different colors for outliers
sns.stripplot(
    y=statistics_df['total'], 
    hue=statistics_df['is_outlier'],  # Use 'is_outlier' as the hue variable
    palette={True: 'blue', False: 'red'},  # Map colors to True/False
    size=4, 
    dodge=False, 
    legend=False  # Disable legend
)

# Annotate outliers
for i, row in statistics_df[statistics_df['is_outlier']].iterrows():
    plt.annotate(
        row['abbreviation'],  # Label with participant name
        (0, row['total']),   # Position based on total value
        textcoords="offset points",
        xytext=(10, 0),      # Offset label to the right
        ha='center',
        fontsize=10,
        color='blue'
    )

# Step 4: Customize the plot
plt.title('Box Plot of Total Token Count per Participant with Outliers Highlighted (blue)', fontsize=12)
plt.ylabel('Total Token Count', fontsize=10)
plt.xlabel('Participants', fontsize=10)

# Remove x-axis labels since we’re not showing participant labels
plt.xticks([])

# Show the plot
plt.tight_layout()
plt.show()
```
